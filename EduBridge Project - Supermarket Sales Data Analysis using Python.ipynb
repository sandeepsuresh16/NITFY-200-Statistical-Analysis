# Data Visualisation

```python
plt.figure(figsize=(20,8))
sb.lineplot(x='Date',y='Total',data=data,color='g',ci=False)
plt.xticks(rotation='vertical')
plt.title('Plot - Date v/s Number of Sales',size=20)
plt.xlabel('Date',size=15)
plt.ylabel('SALES',size=12)
plt.show()
```

```python
plt.figure(figsize=(20,8))
sb.lineplot(x='Date',y='Total',data=data,color='g',ci=True,hue='Branch',style='Branch')
plt.xticks(rotation='vertical')
plt.title('BRANCH-WISE:  DATE v/s NUMBER OF SALES',size=20)
plt.xlabel('Date',size=15)
plt.ylabel('SALES',size=12)
plt.show()
```

```python
sb.countplot('Product line',data=data)
plt.title('Products Sold(Mean) - Category Wise',size=15)
plt.xlabel('Category',size=15)
plt.ylabel('Count',size=12)
plt.xticks(rotation='vertical')
plt.show()
```

```python
sb.countplot('Payment',data=data)
plt.title('Mean Purchase by Different Payment Modes',size=15)
plt.xlabel('Payment Mode',size=15)
plt.ylabel('Number of Purchases',size=12)
plt.xticks(rotation='vertical')
plt.show()
```

```python
sb.barplot(x='Branch',y="Total",data=data,hue='Product line',ci=False)
plt.title('Branch Wise-Average Sale')
plt.grid()
plt.show()
```

```python
sales=data.groupby('Date')['Total'].sum().reset_index()
sales
```

```python
plt.figure(figsize=(20,5))
sb.lineplot(x='Date',y='Total',data=sales,color='g')
plt.title('Sales Trend')
plt.ylabel('Sales($)')
plt.xticks(rotation='vertical')
plt.show()
```

```python
rev=data.groupby('Branch').sum().reset_index()
rev
```

```python
sb.barplot(x='Branch',y='Total',data=rev)
plt.ylabel("Revenue in $")
plt.title('Total Revene')
plt.show()
```

```python
sb.barplot(x='Branch',y='gross income',data=rev,color='b')
plt.ylabel("Income in $")
plt.title('NET PROFIT')
plt.show()
```

```python
data['hour']=data['Time'].apply(lambda x: x.split(':')[0])
hour=data.groupby('hour')['Total'].sum().reset_index().sort_values(by='hour')
```

```python
plt.figure(figsize=(10,5))
sb.lineplot(x=hour['hour'],y=hour['Total'])
plt.title('Plot Showing Sales from 10am - 8pm')
plt.xlabel('Time(24Hr-Format)')
plt.ylabel('Sales($)')
plt.show()
```

```python
pie=data.groupby('Product line')['Total'].sum().reset_index()
```

```python
slices=list(pie['Total'])
name=list(pie['Product line'])
print(slices, name)
```

```python
plt.pie(slices,labels=name,shadow=True,startangle=90,explode=[0.05,0.05,0.05,0.05,0.05,0.05],autopct='%1.1f%%')
plt.title('REVENUE')
#plt.axis('auto')
plt.show()
```

```python
c_type=data.groupby('Customer type')['Total'].sum().reset_index()
slices1=list(c_type['Total'])
name1=list(c_type['Customer type'])
```

```python
plt.pie(slices1,labels=name1,shadow=True,colors=['pink','red'],explode=[0.1,0],startangle=90,autopct='%1.1f%%')
plt.title('Revenue shares from different customers')
plt.show()
```

```python
sb.distplot(data['Rating'],bins=10,color='r')
plt.title('Histogram showing RATING given by Customers')
plt.grid(color='b')
plt.show()
```

```python
sb.displot(data['gross income'],bins=20)
plt.title('Histogram showing GROSS INCOME per date')
plt.grid()
plt.show()
```

```python
sb.scatterplot(x='Total',y="Tax 5%",data=data)
plt.title('ScatterPlot Showing Distribution of Tax v/s Sale')
plt.xlabel('Sale')
plt.show()
```

```python
sb.scatterplot(x='gross margin percentage', y='Total', data=data)
plt.title(' Plot showing distribution of gross margin % to price')
plt.ylabel('Total Price')
```

```python
sb.regplot(y=data['Tax 5%'],x=data['Total'],color='g')
plt.title('Regression Plot TAX v/s PRICE',size=10)
plt.xlabel('Price')
plt.ylabel('Tax')
plt.show()
```

```python
sb.scatterplot(x='Rating',y='Total',data=data,color='r')
plt.title('ScatterPlot Rating v/s Price')
plt.ylabel('Price')
plt.show()
```
